*PART A: Import Libraries*

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

*PART B: Load and Inspect Dataset*

```python
# Load the Iris dataset
df = pd.read_csv('/Iris.csv')

# Drop 'Id' column as it is not needed
df = df.drop(columns=['Id'])

# Inspect dataset
df.info()
print(df.isnull().sum())
```

*PART C: Data Exploration*

**Histograms of Feature Columns**

```python
# Sepal Length Distribution
df['SepalLengthCm'].hist()
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Sepal Width Distribution
df['SepalWidthCm'].hist()
plt.title('Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()

# Petal Length Distribution
df['PetalLengthCm'].hist()
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Petal Width Distribution
df['PetalWidthCm'].hist()
plt.title('Petal Width Distribution')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

**Scatter Plots**

```python
# Sepal Length vs Sepal Width
colors = ['red', 'green', 'blue']
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'], c=colors[i], label=species[i])

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()

# Petal Length vs Petal Width
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['PetalLengthCm'], x['PetalWidthCm'], c=colors[i], label=species[i])

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend()
plt.show()
```

*PART D: Correlation Matrix*

```python
df_numeric = df.drop('Species', axis=1)
corr = df_numeric.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

*PART E: Label Encoding*

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

*PART F: Train-Test Split*

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Species'])
Y = df['Species']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

*PART G: Logistic Regression Mode*

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(x_train, y_train)
```

*PART H: Model Evaluation*

```python
from sklearn.metrics import precision_score, recall_score, f1_score

y_pred = model.predict(x_test)

print("Accuracy:", model.score(x_test, y_test) * 100)
print("Precision:", precision_score(y_test, y_pred, average='weighted') * 100)
print("Recall:", recall_score(y_test, y_pred, average='weighted') * 100)
print("F1 Score:", f1_score(y_test, y_pred, average='weighted') * 100)
```
