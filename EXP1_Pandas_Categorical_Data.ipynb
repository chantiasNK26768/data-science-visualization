EXPERIMENT 1(a): Ordered Categorical Data



*Cell 1 – Import Library & Input Data*

```python
# Importing the required library
import pandas as pd

# List of months
months = ["March", "April", "January", "December", "February",
          "September", "August", "May", "June", "July",
          "October", "November", "March", "April", "January", "December"]

print("VTU26768 - OUTPUT SUMMARY OF EXP1a")
print("Original Months:")
months
```

*Cell 2 – Create Ordered Categorical Variable*

```python
# Create ordered categorical data
ordered_months = pd.Categorical(
    months,
    categories=["January", "February", "March", "April", "May", "June",
                "July", "August", "September", "October", "November", "December"],
    ordered=True
)

print("Ordered Months:")
ordered_months
```

*Cell 3 – Frequency Table*

```python
# Compute frequency table
print("Value Counts:")
month_counts = pd.Series(ordered_months).value_counts().sort_index()
month_counts
```

EXP-1(a) Conclusion (Markdown)

### Conclusion:
Ordered categorical variables help in maintaining logical order.
Frequency tables help in understanding the distribution of data.


EXPERIMENT 1(b): DataFrame & Statistical Analysis

*Cell 4 – Import Libraries & Create DataFrame*

```python
# Importing required libraries
import pandas as pd
import numpy as np

# Create DataFrame
table = {
    "names": pd.Categorical(["chanti", "Nagendra", "kondra", "siva", "vishnu",
                             "jayaprakash", "sasi", "naveen", "kishan", "Shannu"]),
    "Age": np.random.randint(15, 50, size=10),
    "Salary": np.random.normal(50000, 10000, size=10),
    "Categories": pd.Categorical(["HR", "IT", "Finance", "IT", "Marketing",
                                  "IT", "Finance", "IT", "Marketing", "Finance"])
}

dataframe = pd.DataFrame(table)

print("VTU26768 - OUTPUT SUMMARY OF EXP1b")
print("DataFrame:")
dataframe
```

*Cell 5 – Statistical Summary*

```python
# Statistical summary
print("Statistical Summary:")
dataframe.describe(include='all')
```


*Cell 6 – Data Types*

```python
# Display datatypes
print("Datatypes of the table:")
dataframe.dtypes
```

*Cell 7 – DataFrame Information*

```python
# Display DataFrame info
print("Information about the table:")
dataframe.info()
```

EXP-1(b) Conclusion (Markdown)
### Conclusion:
Pandas DataFrame functions like `describe()`, `dtypes`, and `info()`
help in understanding structure, statistics, and nature of the data.
